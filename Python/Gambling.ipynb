```python
import random
import matplotlib.pyplot as pt
x=[]
y=[]
account=0
for i in range(365):
    x.append(i+1)
    bet=random.randint(1,10)#int(input("Enter your bet value:"))
    result=random.randint(1,10)
    if result==bet:
        account+=900-100
    else:
        account-=100
    y.append(account)
print("Final account:",account)
pt.plot(x,y)
pt.show()
    

```
