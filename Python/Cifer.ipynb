```python
import string
dist={}
file=open("test_file.txt","w")
for i in range(len(string.ascii_letters)):
    dist[string.ascii_letters[i]]=string.ascii_letters[i-2]
with open("open_test_file.txt") as f:
    while True:
        c=f.read(1)
        if not c:
            print("End of file")
            break
        print(c)
        if c in dist:
            data=dist[c]
        else:
            data=c
        file.write(data)
file.close()
    
```
